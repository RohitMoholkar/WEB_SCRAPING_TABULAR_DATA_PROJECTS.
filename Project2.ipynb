```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
URL = 'https://www.jagranjosh.com/general-knowledge/asian-games-2023-medal-tally-country-wise-medals-and-winners-list-1695641256-1' 
```

```python
info = BeautifulSoup(response.content) 
```

```python
info 
```

```python
table_info = info.find('table') 
```

```python
table_info
```

```python
tr_val = table_info.find('tr') 
```

```python
tr_val
```

```python
td_val = tr_val.find_all('td') 
```

```python
td_val
```

```python
title_list = []

for val in td_val:
    
    title_list.append(val.text) 

title_list
```

```python
DataFrame = pd.DataFrame(columns=['Rank', 'Country', 'Gold', 'Silver', 'Bronze', 'Total']) 
```

```python
DataFrame
```

```python
tr_tags = table_info.find_all('tr') 
```

```python
tr_tags
```

```python
data_list = [] 

for val in tr_tags[1:]:
    
    data_list = [] 
    
    data = val.find_all('td') 
    
    for td_val in data:
        
        data_list.append(td_val.text.strip()) 
    
    len_df = len(DataFrame) 
    
    DataFrame.loc[len_df] = data_list 
```

```python
DataFrame
```

```python
DataFrame.set_index('Rank' ,inplace=True)
```
