```python
info = BeautifulSoup(response.content, 'html.parser') 
```

```python
info
```

```python
table = info.find('table', class_='ssrcss-krgddh-TableWrapper etjjy0b5')  
```

```python
table 
```

```python
temp_titles = table.find('thead', class_='ssrcss-mtrp4k-TableHead etjjy0b4') 
```

```python
titles = []

for val in temp_titles:
    
    titles.append(val.text) 
```

```python
titles
```

```python
DataFrame = pd.DataFrame(columns=['Rank', 'Country', 'Gold', 'Silver', 'Bronze', 'Total']) 
```

```python
DataFrame
```

```python
temp_data = table.find_all('tr') 
```

```python
for val in temp_data[1:]:
    
    temp_val = []
    
    data = val.find_all('td') 
    
    for i in data:
        
        temp_val.append(i.text) 
    
    l = len(DataFrame) 
    DataFrame.loc[l] = temp_val 
```
