```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
URL = 'https://www.cricketworld.com/cricket/series/icc-cricket-world-cup-qualifiers-2023/stats/batting-most-runs/127711' 
```

```python
response = requests.get(URL) 
```

```python
info = BeautifulSoup(response.content, 'html.parser') 
```

```python
info
```

```python
info.prettify() 
```

```python
table_info = info.find('table', class_='rankingTable batting_highest_strikerate') 
```

```python
table_info
```

```python
temp_list = table_info.find('tr', class_='tableHeader') 
```

```python
temp_list
```

```python
headers = temp_list.find_all('th') 
```

```python
headers 
```

```python
headers_list = []

for val in headers:
    
    headers_list.append(val.text.strip()) 

headers_list
```

```python
DataFrame = pd.DataFrame(columns=['Player','MAT','INNS','NO','RUNS','HS','AVG','BF','SR','100s','50s','6s','4s']) 
```

```python
DataFrame
```

```python
temp_list = info.find_all('tr', class_='statRow format-firstclass') 
```

```python
temp_list
```

```python
for val in temp_list:  
    
    data_list = [] 
    
    data = val.find_all('td') 
    
    for x in data:
        
        data_list.append(x.text.strip()) 
    
    print(data_list) 
    
    len_df = len(DataFrame) 
    
    DataFrame.loc[len_df] = data_list
```
