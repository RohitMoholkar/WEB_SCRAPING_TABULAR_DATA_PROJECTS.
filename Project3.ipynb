```python
for val in driver.find_elements(By.XPATH, '//div[@class="indiright_tbl"]//tbody'): 
    
    temp_str = val 
```

```python
data = temp_str.text.split('\n') 
```

```python
data
```

```python
table_data = [] 
count = 2

for val in data:
    
    if count%2==0:
        
        table_data.append(val) 
    
    count = count + 1

table_data
```

```python
name = [] 
LTP = [] 
change_percent = [] 
volume = [] 
buy_price = [] 
sell_price = [] 
buy_qty = [] 
sell_qty = [] 

for val in table_data:
    
    temp_str1 = ""
    temp_str2 = "" 
    count = 0 
    new_list = [] 
    
    for ch in val: 
        
        if ch>='0' and ch<='9':
            
            break 
        
        temp_str1 = temp_str1 + ch
        
        count = count + 1
    
    name.append(temp_str1) 
    
    temp_str2 = val[count:] 
    
    new_list = temp_str2.split() 
    
    LTP.append(new_list[0]) 
    change_percent.append(new_list[1])
    volume.append(new_list[2]) 
    buy_price.append(new_list[3]) 
    sell_price.append(new_list[4]) 
    buy_qty.append(new_list[5])
    sell_qty.append(new_list[6]) 
```
